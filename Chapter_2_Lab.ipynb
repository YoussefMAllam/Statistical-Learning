```python
import numpy as np
from matplotlib.pyplot import subplots
rng=np.random.default_rng(1303)

x=rng.standard_normal(100)
y=rng.standard_normal(100)

fig, axes=subplots(nrows=2, ncols=3, figsize=(15,10))
axes[0,0].plot(x,y);
axes[0,0].set_xlabel('X-axis for first Graph')
axes[0,0].set_ylabel('Y-axis for first Graph')
axes[0,0].set_title('Line')

axes[0,1].plot(x,y);
axes[0,1].set_xlabel('X-axis for Sccond Graph')
axes[0,1].set_ylabel('Y-axis for Second Graph')
axes[0,1].set_title('Scatter')

axes[0,2].plot(x,y);
axes[0,2].set_xlabel('X-axis for Third Graph')
axes[0,2].set_ylabel('Y-axis for Third Graph')
axes[0,2].set_title('Plus')


axes[1,0].plot(x,y);
axes[1,0].set_xlabel('X-axis for first Graph')
axes[1,0].set_ylabel('Y-axis for first Graph')
axes[1,0].set_xlim([-5,5])
axes[1,0].set_title('Line')
```

```python
import numpy as np
from matplotlib.pyplot import subplots
rng=np.random.default_rng(1303)

fig,axes= subplots(nrows=2, ncols=1,figsize=(8,8))
x=np.linspace(-np.pi,np.pi,50)
y=x
f=np.multiply.outer(np.cos(y),1/(1+x**2))
axes[0].contour(x,y,f)
axes[1].imshow(f)
```
